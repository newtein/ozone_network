```python
import seaborn as sns
import matplotlib.pyplot as plt
from collections import OrderedDict
import numpy as np
import pickle
```

```python
yearMap=OrderedDict()
yearMapRev=OrderedDict()
count=0
for year in range(1980,2018):
    yearMap[str(year)]=count
    yearMapRev[count]=str(year)
    count+=1
```

```python
stateCodes=pickle.load(open("stateCode.pickle","rb"))
paths=[str(i)+'/' for i in stateCodes]
files=['correlationDistribution0.5','correlationDistribution0.7','correlationDistribution0.9']

data=OrderedDict()
for path in paths:
    if path=='16/':
                print("ok")
    for file in files:
        key=file[-3:]
        if key not in data:
            data[key]=OrderedDict()
            
        f=open(path+file+".csv","r")
        for l in f:
            row=l.strip().split(",")
            n=float(row[-2])
            #if n>=10:
            if row[-1]!='None':
                percent=float(row[-1])
            else:
                percent=-1.0
            year=row[0].split('_')[-1]
            key2=path[:-1]
            if key2=='16':
                print("ok2")
            if key2 not in data[key]:
                data[key][key2]=OrderedDict()
            if percent!=-1:
                data[key][key2][yearMap[year]]=percent
            if key2=='16':
                print("ok3")
```
