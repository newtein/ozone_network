```python
import numpy as np
import matplotlib.pyplot as plt
import string
import matplotlib.gridspec as gridspec
import matplotlib.image as mpimg
```

```python
path='Com/images/Figure1'
```

```python
def readImg(fname):
    return mpimg.imread(fname)


```

```python
plt.close()

ax = plt.subplot(111) 
ax.imshow(readImg('{}/OzoneNetwork.png'.format(path)))
ax.text(-0.01, 1, 'e)', transform=ax.transAxes, size=10,color='purple')
plt.axis('off')
plt.savefig("Com/images/Figure1/OzoneNetwork2.png",bbox_inches='tight',dpi=300)
plt.show()
```

```python
n=0
plt.close()
plt.figure(figsize = (8,5), dpi = 300)
gs = gridspec.GridSpec(2,4)

ax=plt.subplot(gs[0,0:4])
ax.imshow(readImg('{}/results.png'.format(path)))  
ax.text(-0.01, 1.0, string.ascii_lowercase[n], transform=ax.transAxes, size=10)
plt.axis('off')
n+=1


ax=plt.subplot(gs[1,0:2])
ax.imshow(readImg('{}/clusteringcoefficient.png'.format(path)))  
ax.text(-0.01, 1.0, string.ascii_lowercase[n], transform=ax.transAxes, size=10)
plt.axis('off')
n+=1

ax=plt.subplot(gs[1,2:4])
ax.imshow(readImg('{}/correlationDistribution.png'.format(path)))  
ax.text(-0.01, 1.0, string.ascii_lowercase[n], transform=ax.transAxes, size=10)
plt.axis('off')

plt.show()
```
