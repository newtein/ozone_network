```python
import pickle
```

```python
targetDict={"NA/1":"1987-1989","NA/2":"2001-2003","NA/3":"2008-2010","NA/4":"2014-2016"}


def standard1979(x):
    if x>=0.280:
        return "Extreme"
    elif 0.190<=x<0.280:
        return "Severe 17"
    elif 0.180<=x<0.190:
        return "Severe 15"
    elif 0.160<=x<0.180:
        return "Serious"
    elif 0.138<=x<0.160:
        return "Moderate"
    elif 0.121<=x<0.138:
        return "Marginal"
    elif x<0.121:
        return "Submarginal"
    else:
        return "Incomplete"
    
def standard2015(x):
    if x>=0.163:
        return "Extreme"
    elif 0.111<=x<0.163:
        return "Severe 17"
    elif 0.105<=x<0.111:
        return "Severe 15"
    elif 0.093<=x<0.105:
        return "Serious"
    elif 0.081<=x<0.093:
        return "Moderate"
    elif 0.071<=x<0.081:
        return "Marginal"

def getLabel(target,x):
    if target=='1987-1989':
        return standard1979(x)
    else: 
        return standard2015(x)
        
```

```python

for targetWrite in targetDict:
    target=targetDict[targetWrite]
    fcode=open('NA/files/NA{}.tsv'.format(target),'r')
    dataNA={}
    for l in fcode:
        row=l.strip().split('\t')
        key=row[0]
        if row[1]!='':
    
            dataNA[key]=[float(row[1]),row[2]]
        else:
            dataNA[key]=[-100,row[2]]
    fcode.close()
    
    f=open('NA/files/NA{}Codes.tsv'.format(target),'r')
    data={}
    ss=0
    for l in f:
        row=l.strip().split('\t')
        key=row[2]
        ss+=1

        if dataNA[key][1]==target:
            if key not in data:
                data[key]=[]
            data[key].append([row[1],dataNA[key][0],row[-1]])
        else:
            print("?",row,dataNA[key])
    print(ss)
    f.close()
    ss=0
    for i in data:
        for j in data[i]:
            ss+=1
    print(ss)
    dataShrink={}
    for region in data:
        for county in data[region]:
            temp=[county[2],county[1]]
            key2=county[2]
            if county[1]!=-100:
                if key2 not in dataShrink:
                    dataShrink[key2]=[]
                dataShrink[key2].append(county[1])
            
    for region in data:
        for irange in range(len(data[region])):
            if data[region][irange][1]!=-100:
                data[region][irange][1]=getLabel(target,data[region][irange][1])
            else:
                del data[region][irange]
            
    for countyCodes in dataShrink:
        dataShrink[countyCodes]=getLabel(target,max(dataShrink[countyCodes]))
        
    print('[]',len(data),len(dataShrink))
    
    pickle.dump(data,open('{}/NAData.pickle'.format(targetWrite),'wb'))
    pickle.dump(dataShrink,open('{}/NAShrink.pickle'.format(targetWrite),'wb'))
    print(target)
```
