```python
import csv
from collections import OrderedDict
import datetime 
import os
import pickle
```

```python
start_date = datetime .date(1998, 1, 1)
end_date = datetime .date(2017,12,31)
day_count = (end_date - start_date).days + 1
def init_date():
    for single_date in (start_date + datetime.timedelta(n) for n in range(day_count)):
        yield single_date.strftime("%Y-%m-%d")
dates = init_date()        
```

```python
fname = open('energy/energy/hourly-energy-consumption/pjm_hourly_est.csv','r')
header = next(fname)
print(header)
data = OrderedDict()
reader =csv.reader(fname)
for row in reader:
    date,hour,loads = row[0].split()[0],row[0].split()[1],row[1:]
    load = sum([float(i) for i in loads if i!=''])
    if date not in data:
        data[date] = []
    if load:
        data[date].append(load)
for date in data:
    data[date] = sum(map(float,data[date]))  
```

```python
wrapper = OrderedDict()
for date in dates:
    year = date.split('-')[0]
    if year not in wrapper:
        wrapper[year] = []
    load = data.get(date,'-')
    wrapper[year].append(load)
```

```python
for year in wrapper:
    path = 'D:/Archived/Cancer/Data/PMJ/{}'.format(year)
    try:
        os.mkdir(path)
    except:
        pass
    print(year,len(wrapper[year]))
    pickle.dump(wrapper[year],open(path+'/load.pickle','wb'))
```
