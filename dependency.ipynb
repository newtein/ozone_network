```python
import os
import pickle
from collections import OrderedDict
import commonFunctions as cf
```

```python
stateCodes=pickle.load(open("stateCode.pickle","rb"))

stateCodes=list(stateCodes.keys())
```

```python
import scipy
```

```python
var={"o3":"44201","TEMP":"TEMP"}
variables=list(var.keys())
startYear=1980
endYear=1981
data=OrderedDict()
for year in range(startYear,endYear):
    if year not in data:
        data[year]=OrderedDict()
    usaData=[[],[]]
    for stateCode in stateCodes:
        path=str(stateCode) + "/Data"
        if stateCode not in data[year]:
            data[year][stateCode]=OrderedDict()
      
       
        path1=path+"/"+variables[0]+"/"+str(year)
        path2=path+"/"+variables[1]+"/"+str(year)
        countyList=set(os.listdir(path1)).intersection(set(os.listdir(path2)))
        for county in countyList:
            county1Data=pickle.load(open(path1+"/"+county,"rb"))
            county2Data=pickle.load(open(path2+"/"+county,"rb"))
            usaData[0].extend(county1Data)
            usaData[1].extend(county2Data)
            
            countyName=county.replace(".pickle","")
            if countyName not in data[year][stateCode]:
                data[year][stateCode][countyName]=(-1,-1)
            data[year][stateCode][countyName]=cf.correlation(county1Data,county2Data,gaps=True)
                
                    
            
            
    
        
        
        
    
```
