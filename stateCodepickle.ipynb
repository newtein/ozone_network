```python
import pickle
from collections import OrderedDict
```

```python
data=OrderedDict()
f=open("states_and_counties.csv","r")
next(f)
for l in f:
    row=l.strip().split(',')
    
    stateCode=int(row[0].replace('"',''))
    print(stateCode)
    stateName=row[1]
    if stateName not in data:
        data[stateCode]=stateName
```

```python
pickle.dump(data,open("stateCode.pickle","wb"))
```
